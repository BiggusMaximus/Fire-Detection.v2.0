```python
import cv2
import glob, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_ori = cv2.imread('./dataset/all/obj/14.jpg')
img = cv2.cvtColor(img_ori,  cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Grayscale

```python
img_grayscale = cv2.cvtColor(img_ori, cv2.COLOR_BGR2GRAY)
img_grayscale = cv2.cvtColor(img_grayscale,  cv2.COLOR_BGR2RGB)
plt.imshow(img_grayscale)
```

# Erosion

```python
kernel = np.ones((5, 5), np.uint8)
img_erosion = cv2.erode(img_grayscale, kernel, iterations=3)
plt.imshow(img_erosion)
```
